## How to get total number of transfer events for a token


```python
token_contract_address = '0x1F573D6Fb3F13d689FF844B4cE37794d79a7FF1C'

# Standard ERC-20 Transfer event signature
transfer_event_signature_hash = w3.keccak(text="Transfer(address,address,uint256)").hex()
```

```python
def fetch_transfer_events(from_block, to_block):
    event_filter = {
        'address': token_contract_address,
        'fromBlock': from_block,
        'toBlock': to_block,
        'topics': [transfer_event_signature_hash]
    }
    return w3.eth.get_logs(event_filter)
```

```python
# start_block = 0  # Starting block for your query, adjust based on the token's deployment block
# end_block = w3.eth.block_number  # Latest block, adjust if querying historical data
start_block = 0x3AC45B
end_block = 0x3AEBBF
block_step = 500000  # Adjust based on network activity and response size

total_transfers = 0
for block in range(start_block, end_block, block_step):
    events = fetch_transfer_events(block, min(block + block_step, end_block))
    total_transfers += len(events)
    print(f"Fetched {len(events)} events from blocks {block} to {min(block + block_step, end_block)}")

print(f"Total Transfer events: {total_transfers}")
```
